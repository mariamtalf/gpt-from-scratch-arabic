```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace

file_path = "../data/pretrain/data.txt"

tokenizer = Tokenizer(BPE(unk_token="[UNK]"))
tokenizer.pre_tokenizer = Whitespace()

trainer = BpeTrainer(
    vocab_size=5000,
    min_frequency=2,
    special_tokens=["[PAD]", "[UNK]", "[BOS]", "[EOS]"]
)

tokenizer.train([file_path], trainer)
tokenizer.save("../src/tokenizer/arabic_bpe_tokenizer.json")
```

```python
vocab = tokenizer.get_vocab()
print("Vocabulary size:", len(vocab))
```

```python
sample_text = "كانت نهاية الأسرة الثامنة عشرة — وهي آخر مرحلة وصلنا إليها في الجزء السالف"
encoded = tokenizer.encode(sample_text)

print("Original text:")
print(sample_text)

print("\nTokens:")
print(encoded.tokens)

print("\nIDs:")
print(encoded.ids)

decoded = tokenizer.decode(encoded.ids)
print("\nDecoded text:")
print(decoded)
```

```python
sample_text2 = "ذهب الطفل إلى المدرسة في الصباح الباكر."
encoded2 = tokenizer.encode(sample_text2)

print("Original text:")
print(sample_text2)

print("\nTokens:")
print(encoded2.tokens)

print("\nDecoded text:")
print(tokenizer.decode(encoded2.ids))
```
